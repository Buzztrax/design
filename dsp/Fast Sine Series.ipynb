```python
angle = 2 * pi / 100;
l = 400
```

Calculate a buffer of sine values.

```python
p = 0
v = []
for i in range(l):
    v.append(sin(p))
    p += angle

plot(v)
grid(True)
show()
print(v[0], v[-1])
```

Fast version that caclulates followup values using one multiplication and one subtraction

```python
si0 = sin(-angle)
si1 = 0.0
fc = 2.0 * cos(angle)
v = []
v.append(si1)
for i in range(l//2 - 1):
    si0 = fc * si1 - si0
    v.append(si0)
    si1 = fc * si0 - si1
    v.append(si1)

plot(v)
grid(True)
show()
print(v[0], v[-1])
```

```python
sk = 0.0
ck = 1.0
s1 = sin(angle) 
c1 = cos(angle)
v = []
for i in range(l):
    skk = c1 * sk + s1 * ck
    ckk = c1 * ck - s1 * sk
    sk = skk
    ck = ckk
    v.append(sk)

plot(v)
grid(True)
show()
print(v[0], v[-1])
```

Same as before but demonstrating how initializie the series with a phase != 0

```python
phase = pi / 2.0
si0 = sin(phase-angle)
si1 = sin(phase)
fc = 2.0 * cos(angle)
v = []
v.append(si1)
for i in range(l//2 - 1):
    si0 = fc * si1 - si0
    v.append(si0)
    si1 = fc * si0 - si1
    v.append(si1)

plot(v)
grid(True)
show()
print(v[0], v[-1])
```
