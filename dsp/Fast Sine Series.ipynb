```python
angle = 2 * pi / 100;
l = 400
```

```python
p = 0
v = []
for i in range(l):
    v.append(sin(p))
    p += angle

plot(v)
grid('on')
show()
print v[0], v[-1]
```

```python
si0 = sin(-angle)
si1 = 0.0
fc = 2.0 * cos(angle)
v = []
v.append(si1)
for i in range(l/2 - 1):
    si0 = fc * si1 - si0
    v.append(si0)
    si1 = fc * si0 - si1
    v.append(si1)

plot(v)
grid('on')
ylim([-1.,1.])
show()
print v[0], v[-1]
```
